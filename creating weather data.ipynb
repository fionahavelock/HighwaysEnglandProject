```
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Point, Daily
import pandas as pd

# Set time period
start = datetime(1979, 1, 1)
end = datetime(2021, 12, 31)


# Create Point for Vancouver, BC
london= Point(51.489096, -0.191170, 45)

# Get daily data for 2018
data = Daily(london, start, end)
data = data.fetch()


# Plot line chart including average, minimum and maximum temperature
data.plot(y=["tavg", "tmin", "tmax"])
plt.show()
```

```
df = pd.DataFrame(data)
```

```
# snow: The snow depth in mm
# prcp: The daily precipitation total in mm
# tavg: The average air temperature in °C
# wspd: The average wind speed in km/h
# time: The date
display(df)
```

```
df['snow'] = df['snow'].fillna(0)
df['prcp'].fillna(method='ffill', inplace=True)
df['tavg'].fillna(method='ffill', inplace=True)
df['wspd'].fillna(method='ffill', inplace=True)
df['time'] = df.index
df = df.drop(['tmin', 'tmax', 'wdir', 'wpgt', 'pres', 'tsun'], axis=1)
```

```
df.rename(columns = {'snow':'snowDepth', 'prcp':'dailyRain','tavg':'averageTemp', 'wspd':'windSpeed', 'time':'day'}, inplace = True)
display(df)
```

```
# from  pyspark library import 
# SparkSession
from pyspark.sql import SparkSession
  
# Building the SparkSession and name
# it :'pandas to spark'
spark = SparkSession.builder.appName(
  "pandas to spark").getOrCreate()
df_spark = spark.createDataFrame(df)
```

```
filepath = "/FileStore/tables/historicalWeatherData"
df_spark.write.option("header","true").option("sep",",").mode("overwrite").csv(filepath)
```
