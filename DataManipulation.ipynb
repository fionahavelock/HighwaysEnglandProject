```python
months = []

for ele in LondonAccidents['date'].values:
    month = ele.split('/')
    months.append(int(month[1]))
    #print(hours)
    
LondonAccidents["month"] = months
LondonAccidents = LondonAccidents.drop(['date'], axis=1)
```
